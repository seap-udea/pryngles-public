## Exoring in the new Pryngles

Add a ring:

```python
sys=System()
S=sys.add()
P=sys.add("Planet",primary=S,radius=0.1,x=3)
R=sys.add("Ring",primary=P,
          fi=1.3,fe=2.3,
          i=-20*Consts.deg,roll=0*Consts.deg)
sys.spangle_system()
sys.update_perspective(n_obs=Science.direction(120,0))
sys.sg.plot2d(include=["Planet","Ring"],show_azim=False);
```

Special variable with transit conditions:

```python
cond=(sys.data.transit_over_obs!="")
sys.data.loc[cond,["transit_over_obs"]]
```

Highlights points in planet seing through rings:

```python
cond=(sys.data.transit_over_obs.apply(lambda x:'Planet' in x))
sys.sg.plot2d(include=["Planet","Ring"],highlight=(cond,dict()))
```

Highlight points in partial shadow:

```python
cond=(sys.data.hidden_by_luz.apply(lambda x:'Ring' in x))&\
     (sys.data.visible)
sys.sg.plot2d(include=["Planet","Ring"],highlight=(cond,dict()));
```

## Exomoons in the new Pryngles

System with an exomoon:

```python
sys=System()

S=sys.add()
P=sys.add("Planet",primary=S,radius=0.1,x=3)
R=sys.add("Ring",primary=P,
          fi=1.3,fe=2.3,
          i=-20*Consts.deg,roll=0*Consts.deg)

M=sys.add("Planet",bhash="Moon",primary=P,radius=0.02,x=2.5,z=0.05)

sys.spangle_system()
sys.update_perspective(n_obs=Science.direction(120,0))
sys.sg.plot2d(include=["Planet","Moon","Ring"],show_azim=False)
```

Check orbit:

```python
Plot.animate_rebound(sys.sim,filename="figures/system-animate.gif")
Image(open("figures/system-animate.gif","rb").read())
```

Create an orbit for a hierarchical N-body system:

```python
units=['au','msun','yr']
orbit=Orbit(m1=1,
            m2=Orbit(m1=1e-3,
                     m2=1e-6,
                     units=units,
                     a=0.5,inc=30*Consts.deg,M=180*Consts.deg),
            units=units,
            a=20,e=0)
orbit.calculate_orbit()
Plot.animate_rebound(orbit.sim,filename="figures/system-moon-animation.gif")
Image(open("figures/system-moon-animation.gif","rb").read())
```

Use that orbit for creating Planetary System:

```python
units=['au','msun','yr']
orbit=Orbit(m1=1,
            m2=Orbit(m1=1e-3,
                     m2=1e-6,
                     units=units,
                     a=0.5,inc=30*Consts.deg,M=180*Consts.deg),
            units=units,
            a=20,e=0)
orbit.calculate_orbit()
sim,states=orbit.get_states()

sys=System(units=units)

S=sys.add(**states[0])
P=sys.add("Planet",primary=S,radius=0.1,**states[1])
R=sys.add("Ring",primary=P,
          fi=1.3,fe=2.3,
          i=-20*Consts.deg,roll=0*Consts.deg)

M=sys.add("Planet",bhash="Moon",primary=P,radius=0.02,**states[2])

sys.spangle_system()
sys.update_perspective(n_obs=Science.direction(120,0))
sys.sg.plot2d(include=["Planet","Moon","Ring"],show_azim=False)
```

Integrate the system:

```python
sys.integrate(6)
sys.update_perspective()
sys.sg.plot2d(include=["Planet","Moon","Ring"]);
```

Animate the visualization:

```python
from tqdm import tqdm
for i,t in enumerate(tqdm(np.linspace(0,20,50))):
    sys.integrate(t)
    sys.update_perspective()
    if i==0:
        del sys.sg.fig2d
    sys.sg.plot2d(include=["Planet","Moon","Ring"],newfig=False)
    if i==0:
        camera=Camera(sys.sg.fig2d)
    else:
        camera.snap()
```

Save animation:

```python
anim=camera.animate(interval=50,repeat=True,blit=False)
anim.save("figures/planet-moon-animation.gif")
Image(open("figures/planet-moon-animation.gif","rb").read())
```
